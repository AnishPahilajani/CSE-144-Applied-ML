## Tensor Math and comparision oporations

```python
x = torch.tensor([1,2,3])
y = torch.tensor([9,8,7])

# Addition
z1 = torch.empty(3)
z1 = torch.add(x, y)
z1
```

```python
x+y
```

```python
x-y
```

```python
x/y
```

```python
t = torch.zeros(3)
t.add_(x) # underscre represents inplace operation
```

```python
t += x
t
```

```python
#Exponentiaiton
x.pow(2) #Elemnt wise
x**2
```

```python
z = x>1
z
```

```python
# matrix multiplicaton
x1 = torch.rand((2,5))
x2 = torch.rand((5,3))
x3 = torch.mm(x1, x2)
x3
```

```python
matrix_exp = torch.rand(5, 5)
matrix_exp.matrix_power(3)
```

```python
# element wise mult
x * y
```

```python
# dot product: Element wise and the sumation of all the values so 9+12+21
torch.dot(x, y)
```
